```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df=pd.read_csv('https://raw.githubusercontent.com/overtunned/DataScience/main/car%20data.csv')
df.head()
```

```python
#Checking  number ofRow and column
df.shape
```

```python
#check missing values or null values
df.isnull().sum()
```

```python
df.describe()
```

```python
df.columns
```

```python
#creating a new dataframe by removing car name
final_dataset=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
final_dataset.head()
```

```python
#creating a new column current year
final_dataset['Current_Year']=2020
```

```python
final_dataset.head()
```

```python
#creating a new column no_year from current year and year
final_dataset['no_year']=final_dataset['Current_Year']-final_dataset['Year']
```

```python
final_dataset.head()
```

```python
final_dataset.columns
```

```python
#dropping current year and year column
final_dataset.drop(["Year","Current_Year"], axis=1, inplace=True)
final_dataset.head()
```

```python
#visualizing vehicle of different fuel type
print(" VISUALIZING COUNT OF FUEL TYPE ")
plt.figure(figsize=(5,5))
sns.countplot(x='Fuel_Type',data=df)
plt.title('Fuel_Type')
print(df['Fuel_Type'].value_counts())
```

```python
#Visualizing number of owners
print(" VISUALISING NO OF OWNERS ")
plt.figure(figsize=(5,5))
sns.countplot(x='Owner',data=df)
plt.title('Owner')
print(df['Owner'].value_counts())
```

```python
#visualizing transmission type of vehicle
print(" Transmission type ")
plt.figure(figsize=(5,5))
sns.countplot(x='Transmission',data=df)
plt.title('Transmission type')
print(df['Transmission'].value_counts())
```

```python
#count of vehicle of different ages
print(" Age of Vehicle ")
plt.figure(figsize=(5,5))
sns.countplot(x='no_year',data=final_dataset)
plt.title('Age of Cars')
print(final_dataset['no_year'].value_counts())
```

```python
#selling price,km driven and present price count
plt.figure(figsize = (16,5))


plt.subplot(1, 3, 1)
plt.hist(df['Selling_Price'])
plt.xlabel('Selling_Price')
plt.ylabel('Count')

plt.subplot(1, 3, 2)
plt.hist(df['Kms_Driven'])
plt.xlabel('Kms_Driven')
plt.ylabel('Count')


plt.subplot(1, 3, 3)
plt.hist(df['Present_Price'])
plt.xlabel('Present_Price')
plt.ylabel('Count')
plt.show()
```

```python
#plotting between present price and selling price
df.plot.scatter(x="Present_Price",y="Selling_Price")
```

```python
#plotting between kms driven and selling price
df.plot.scatter(x="Kms_Driven",y="Selling_Price")
```
